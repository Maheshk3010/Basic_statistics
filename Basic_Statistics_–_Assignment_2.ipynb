```python
import numpy as np
from scipy import stats
```

```python
data = np.array([
    1.13, 1.55, 1.43, 0.92, 1.25,
    1.36, 1.32, 0.85, 1.07, 1.48,
    1.20, 1.33, 1.18, 1.22, 1.29
])

data
```

```python
mean = np.mean(data)
std = np.std(data, ddof=1)   # sample standard deviation

mean, std
```

The sample mean and sample standard deviation are calculated from the given data.


```python
n = len(data)
n
```

```python
t_value = stats.t.ppf(0.995, n-1)

ci_sample_sd = (
    mean - t_value * std / np.sqrt(n),
    mean + t_value * std / np.sqrt(n)
)

ci_sample_sd
```

Since the population standard deviation is unknown and sample size is small,
t-distribution is used to construct the 99% confidence interval.


```python
sigma = 0.2
z_value = stats.norm.ppf(0.995)

ci_population_sd = (
    mean - z_value * sigma / np.sqrt(n),
    mean + z_value * sigma / np.sqrt(n)
)

ci_population_sd
```

Since the population standard deviation is known, Z-distribution is used.
